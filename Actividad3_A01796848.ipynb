# 1. Introducción teórica

## 1.1 Big Data y procesamiento distribuido

El término **Big Data** hace referencia a conjuntos de datos cuya magnitud, velocidad de generación y variedad exceden las capacidades de procesamiento de los sistemas tradicionales. Se caracteriza por *5 Vs*: volumen, velocidad, variedad, veracidad y valor. Para procesar eficientemente estos datos se utilizan arquitecturas distribuidas, donde tanto el almacenamiento como el procesamiento se reparten entre múltiples nodos de un clúster.

**Apache Spark** es uno de los frameworks más utilizados para el procesamiento distribuido debido a su capacidad de ejecutar operaciones en memoria, reduciendo significativamente los tiempos de procesamiento frente a enfoques tradicionales como MapReduce. Spark proporciona estructuras de datos distribuidas como los *Resilient Distributed Datasets* (RDDs) y los *DataFrames*, además de bibliotecas especializadas para análisis de datos y aprendizaje automático.

## 1.2 Machine Learning

El **Machine Learning (ML)** llamado en español **aprendizaje automático** es una rama de la inteligencia artificial que desarrolla algoritmos capaces de aprender patrones a partir de los datos para realizar predicciones o descubrir estructuras ocultas sin necesidad de programar explícitamente todas las reglas de decisión.

Integrada de manera nativa en Apache Spark tenemos la biblioteca **MLlib**, diseñada para entrenar modelos de aprendizaje automático de manera distribuida sobre grandes volúmenes de información, permitiendo escalar tanto los procesos de entrenamiento como de inferencia [2].

## 1.3 Preprocesamiento y análisis descriptivo de datos

Previo a entrenar un modelo de aprendizaje automático es necesario realizar procesos de **limpieza, transformación y exploración de datos** para garantizar la calidad de los datos y mejorar el desempeño de los modelos de aprendizaje automático. Como por ejemplo manejo de valores faltantes, codificación de variables categóricas, normalización o escalamiento de variables numéricas, generación de nuevas características (*feature engineering*) y análisis estadístico descriptivo de los datos [1].

PySpark proporciona herramientas para realizar estas tareas como:

- **VectorAssembler**: combina múltiples variables en un único vector de características.
- **StringIndexer**: transforma variables categóricas en índices numéricos.
- **OneHotEncoder**: representa variables categóricas mediante vectores binarios.
- **VectorIndexer**: identifica automáticamente variables categóricas dentro de un vector de características.
- **StandardScaler** y **MinMaxScaler**: normaliza variables numéricas.
- **MaxAbsScaler** y **RobustScaler**: escala variables utilizando diferentes estrategias de normalización.
- **Imputer**: reemplaza valores faltantes mediante estadísticas como la media o la mediana.
- **Binarizer**: convierte variables numéricas en variables binarias utilizando un umbral.
- **Bucketizer** y **QuantileDiscretizer**: discretiza variables continuas en intervalos o categorías.
- **PCA (Principal Component Analysis)**: reduce la dimensionalidad de los datos conservando la mayor cantidad posible de información.
- **Tokenizer**, **RegexTokenizer**, **StopWordsRemover**, **CountVectorizer** y **HashingTF**: para el procesamiento y transformación de datos textuales.
- Así como funciones estadísticas para calcular correlaciones, distribuciones, frecuencias y estadísticas descriptivas de los datos [1,4].

## 1.4 Aprendizaje supervisado

El **aprendizaje supervisado** utiliza un conjunto de datos etiquetado, donde cada observación posee una variable objetivo conocida. El objetivo es aprender a predecir dicha variable para nuevas observaciones [2].

Los principales modelos de aprendizaje supervisado son:

- **Clasificación:** predicción de categorías o clases discretas.
- **Regresión:** predicción de valores continuos.

A continuación los algoritmos supervisados disponibles en PySpark MLlib [5]:

#### Clasificación

- Regresión logística: **LogisticRegression**
- Árboles de Decisión: **DecisionTreeClassifier**
- Random Forest: **RandomForestClassifier**
- Gradient Boosted Trees: **GBTClassifier**
- Máquinas de Soporte Vectorial (SVM): **LinearSVC**
- Naive Bayes: **NaiveBayes**
- Redes Neuronales Artificiales: **MultilayerPerceptronClassifier**

También tenemos la herramienta **OneVsRest** que nos sirve para convertir un problema de clasificación multiclase en varios problemas de clasificación binaria.

#### Regresión

- Regresión Lineal: **LinearRegression**
- Regresión Lineal Generalizada: **GeneralizedLinearRegression**
- Árboles de Decisión: **DecisionTreeRegressor**
- Random Forest: **RandomForestRegressor**
- Gradient Boosted Trees: **GBTRegressor**
- Regresión Isotónica: **IsotonicRegression**

## 1.5 Aprendizaje no supervisado

Por otro lado, el **aprendizaje no supervisado** trabaja con datos que no poseen etiquetas conocidas. Su objetivo es descubrir patrones, agrupamientos o estructuras ocultas presentes en los datos [2].

Las principales aplicaciones del aprendizaje no supervisado incluyen:

- Segmentación o agrupamiento (*clustering*).
- Detección de anomalías.
- Reducción de dimensionalidad.
- Descubrimiento de tópicos en colecciones de texto.

A continuación, algoritmos no supervisados disponibles en PySpark MLlib [6]:

- Análisis de Componentes Principales: **PCA**.

#### Modelos probabilísticos y tópicos

PySpark MLlib no tiene un algoritmo específico dedicado exclusivamente a anomalías, pero se construye combinando otros modelos.

### 2.3 Justificación de la selección de variables

Para los modelos se seleccionaron las variables con mayor poder informativo respecto al precio y a las características estructurales de las propiedades: `Room Type`, `Property Type`, `Bedrooms`, `Bathrooms`, `Accommodates`, `Review Scores Rating`, `Availability 365`, `Cleaning Fee` (r=0.55), `Beds` (r=0.29), `Extra People` y `Price`. Esta elección se respalda con el análisis exploratorio previo, en el cual se detectaron correlaciones positivas entre el precio y los atributos de capacidad, además de la relevancia del tipo de habitación como factor diferenciador del segmento de mercado.

### 2.4 Método de muestreo: estratificado

Debido a la magnitud del dataset y al desbalance de algunas categorías, se utilizó un **muestreo estratificado**. Este método consiste en dividir la población en estratos homogéneos respecto a una o varias variables de interés y extraer una fracción de cada estrato, garantizando así la representatividad de todas las subpoblaciones en la muestra final.

En el proyecto previo se construyeron **9 particiones** $P_1, \ldots, P_9$ mediante el cruce de las tres categorías de `Room Type` (`Entire home/apt`, `Private room`, `Shared room`) y las tres categorías de `price_range` (`Low`, `Mid`, `High`). Sobre cada partición se aplicó un muestreo aleatorio simple con fracción del 5% y semilla fija (`seed=42`), obteniendo así submuestras $M_1, \ldots, M_9$, cuya unión conforma la **muestra estratificada M** utilizada. De tal forma que el muestreo estratíficado nos permite:
- Preservar la proporción de las categorías minoritarias (`Shared room` y `price_range = High`), cuya pérdida en un muestreo aleatorio simple sesgaría los modelos hacia las clases mayoritarias.
- Reducir la varianza del estimador al asegurar cobertura completa de los estratos relevantes (estimaciones de desempeño más estables).
- Facilitar el entrenamiento de modelos de clasificación multiclase al mantener la variable objetivo `price_range`, evitando el problema de clases prácticamente ausentes en el conjunto de entrenamiento.

## 3. Preparación de los datos

La preparación de los datos es un paso fundamental en el Machine Learning debido a que la calidad de los datos con que se alimenta un modelo, limita la calidad del modelo. A continuación los pasos de preprocesamiento aplicados.

### 3.1 Eliminación de registros inválidos

Durante la exploración se detectaron valores espurios en `Room Type` (por ejemplo, 9 y valores nulos) que no corresponden a categorías válidas de la plataforma Airbnb. Estos registros fueron eliminados conservando únicamente las categorías válidas, en este caso: `Entire home/apt`, `Private room` y `Shared room`. También se filtraron registros con `Property Type` o `Price` nulo, porque es indispensable tanto para construir la variable objetivo.

### 3.2 Tratamiento de valores nulos

Los valores nulos en variables predictoras numéricas (`Bedrooms`, `Bathrooms`, `Accommodates`, `Review Scores Rating`, `Availability 365`) se imputaron utilizando su respectiva **mediana**. Se eligió la mediana en lugar de la media porque las distribuciones presentan sesgo a la derecha y valores atípicos, lo cual distorsionaría la media. Para variables categóricas (`Property Type`), se reemplazaron por la categoría `"Unknown"`.

### 3.3 Conversión de tipos de datos

Se utilizó `inferSchema=True` resultando en una inferencia inicial razonable, peor fue necesario forzar el tipado de las variables numéricas a `DoubleType` para garantizar la compatibilidad con los transformadores y estimadores de MLlib, los cuales requieren entradas numéricas en precisión doble para el ensamblado del vector de características.

### 3.4 Construcción de la variable categórica price_range

A partir de los cuantiles empíricos del precio ($p_{25} = 55$ USD y $p_{75} = 150$ USD), se discretizó la variable continua `Price` en tres categorías:

- **Low:** $\text{Price} < p_{25}$  
- **Mid:** $p_{25} \leq \text{Price} \leq p_{75}$  
- **High:** $\text{Price} > p_{75}$

Transformando un problema de regresión en uno de clasificación multiclase.

### 3.5 Detección de inconsistencias

Se verificó que `Price`, `Bedrooms`, `Bathrooms` y `Accommodates` no tuvieran valores negativos, descartando registros con valores anómalos. Pro otro lado, se inspeccionaron los dominios de variables categóricas con `distinct()` para detectar codificaciones erróneas.



```python
# Limpieza de datos
df_clean = (
    df.filter(col("Room Type").isin("Entire home/apt", "Private room", "Shared room"))
      .filter(col("Property Type").isNotNull())
      .filter(col("Price").isNotNull())
)

# Variables numéricas
#   Cleaning Fee (r=0.55), Beds (r=0.29), Extra People (r=0.27),
#   Guests Included (r=0.19)
num_cols = ["Price", "Bedrooms", "Bathrooms", "Accommodates", "Beds", "Cleaning Fee", "Extra People", "Guests Included",
    "Number of Reviews", "Host Listings Count","Review Scores Rating", "Availability 365"]
for c in num_cols:
    df_clean = df_clean.withColumn(c, col(c).cast(DoubleType()))

# Filtrar valores imposibles
df_clean = (
    df_clean
    .filter(col("Price") > 0)
    .filter(col("Accommodates") > 0)
    .filter(col("Bedrooms") >= 0)
    .filter(col("Bathrooms") >= 0)
)

# Imputación de nulos numéricos con la mediana
medianas = {c: df_clean.approxQuantile(c, [0.5], 0.01)[0] for c in num_cols}
print("Medianas calculadas:", medianas)
for c, m in medianas.items():
    df_clean = df_clean.fillna({c: m})

# Imputación categórica
df_clean = df_clean.fillna({
    "Property Type": "Unknown",
    "Cancellation Policy": "Unknown",
    "Bed Type": "Unknown"
})

# Feature engineering: log1p(Number of Reviews) para reducir sesgo de larga cola
df_clean = df_clean.withColumn("log_reviews", log1p(col("Number of Reviews")))

print("Registros tras limpieza:", df_clean.count())
```

```python
# Construcción de price_range a partir de cuantiles

p25, p75 = df_clean.approxQuantile("Price", [0.25, 0.75], 0.01)
print(f"p25 = {p25} | p75 = {p75}")

df_clean = df_clean.withColumn(
    "price_range",
    when(col("Price") < p25, "Low")
    .when((col("Price") >= p25) & (col("Price") <= p75), "Mid")
    .otherwise("High")
)

df_clean.groupBy("price_range").count().orderBy("price_range").show()

df_clean = df_clean.withColumn(
    "stratum",
    concat_ws("_",
        col("Room Type"),
        col("price_range")
    )
)

df_clean.groupBy("stratum").count().orderBy("stratum").show(truncate=False)
```

```python
estratos = (
    df_clean
    .select("stratum")
    .distinct()
    .rdd
    .flatMap(lambda x: x)
    .collect()
)

# Fracción aumentada a 10% para mayor representatividad y mejor accuracy
fractions = {e: 0.10 for e in estratos}

muestra_M = df_clean.sampleBy(
    "stratum",
    fractions=fractions,
    seed=42
)

muestra_M = muestra_M.cache()
print("Tamaño muestra M:", muestra_M.count())
```

```python
muestra_M.groupBy("stratum") \
         .count() \
         .orderBy("stratum") \
         .show(truncate=False)
```

### 3.6 Preparación de variables para Machine Learning

Los algoritmos de MLlib no aceptan variables categóricas directamente: requieren una representación numérica vectorial. Por este motivo se aplicaron las siguientes transformaciones:

1. **StringIndexer** sobre `Room Type`, `Property Type` y la variable objetivo `price_range`, convirtiendo cada categoría en un índice entero.
2. **OneHotEncoder** sobre los índices de las predictoras categóricas, para que el modelo no interprete el orden como una relación numérica.
3. **VectorAssembler** para ensamblar las variables predictoras (numéricas y categóricas) en una columna `features` de tipo vectorial.
4. **StandardScaler**  que estandariza las variables a media cero y varianza unitaria, para K-means porque esta basado en distancia euclidiana como K-Means.

## 4. Preparación de los conjuntos de entrenamiento y prueba

El 80% de la muestra M se utiliza en el entrenamiento y el 20% en el conjunto de prueba. Utilizamos la regla 80/20 (principio de Pareto) debido a que en muestras del orden  de decenas de miles, proporciona un balance aceptado en la practica ya que asgina suficientes observadores al entrenamiento para que los estimadores converjan en solucionbes estables y mantiene un conjunto prueba lo suficiente grande para que las métricas de desempeño tengan baja varianza muestral. La partición es pseudo-aleatoria con una semilla fija (`seed=42`) para asegurar la reproducibilidad de los experimentos. Separamos los datos en un conjunto de entrenamiento y uno de prueba para reducir el riesgo de **sobreajuste**, asegurando una estimación más confiable de su capacidad de generalización.

PySpark proporciona el método `randomSplit`, que realiza una partición aleatoria respetando las proporciones indicadas.

```python
# División 80/20 train/test
train_df, test_df = muestra_M.randomSplit([0.8, 0.2], seed=42)

print("Tamaño entrenamiento:", train_df.count())
print("Tamaño prueba:       ", test_df.count())

# Distribución por price_range

print("\nDistribución price_range — Train:")
train_df.groupBy("price_range") \
        .count() \
        .orderBy("price_range") \
        .show()

print("Distribución price_range — Test:")
test_df.groupBy("price_range") \
       .count() \
       .orderBy("price_range") \
       .show()

# Distribución de los 9 estratos

print("\nEstratos Train:")
train_df.groupBy("Room Type", "price_range") \
        .count() \
        .orderBy("Room Type", "price_range") \
        .show(truncate=False)

print("Estratos Test:")
test_df.groupBy("Room Type", "price_range") \
       .count() \
       .orderBy("Room Type", "price_range") \
       .show(truncate=False)
```

## 5. Construcción del modelo de aprendizaje supervisado (Random Forest)

Se eligió **Random Forest Classifier** por las siguientes características:

- **Robustez frente a outliers y escalas heterogéneas**: no requiere normalización previa dbeido a que esta basado en particiones del espacio.
- **Buen manejo de relaciones no lineales**: entre variables predictoras y la variable objetivo.
- **Reducción de varianza por *bagging***: se controla el sobreajuste de los arboles individuales porque promedia múltiples árboles entrenados sobre muestras (bootstrap).
- **Escalabilidad en PySpark**: MLlib implementa Random Forest de forma distribuida.

### 5.1 Variable objetivo y predictoras

- **Variable objetivo:** `price_range` (categórica: Low, Mid, High).
- **Predictoras numéricas:** `Bedrooms`, `Bathrooms`, `Accommodates`, `Review Scores Rating`, `Availability 365`.
- **Predictoras categóricas:** `Room Type`, `Property Type`.

### 5.2 Proceso de indexación, codificación, ensamblado y evaluación

Se creo un pipeline en el siguiente orden:
1. StringIndexer,
2. OneHotEncoder,
3. VectorAssembler,
4. RandomForestClassifier (100 árboles con profundidad máxima 10).

Para la evaluación se utilizaron dos métricas : **Accuracy** para la interpretabilidad y **F1-Score** para la robustez frente al desbalance entre clases.

```python
# Variables categóricas — añadimos Cancellation Policy y Bed Type
cat_cols = ["Room Type", "Property Type", "Cancellation Policy", "Bed Type"]

# Variables numéricas — Cleaning Fee es la feature más correlacionada (r=0.55)
# Se eliminan Review Scores Rating y Availability 365 (r≈0.03, ruido puro)
num_cols_rf = [
    "Bedrooms", "Bathrooms", "Accommodates", "Beds",
    "Cleaning Fee", "Extra People", "Guests Included",
    "log_reviews", "Host Listings Count"
]

# ── Stages de preprocesamiento (compartidos por RF y GBT) ──────────────────
indexers = [
    StringIndexer(inputCol=c, outputCol=c.replace(" ","_")+"_idx", handleInvalid="keep")
    for c in cat_cols
]
encoders = [
    OneHotEncoder(inputCol=c.replace(" ","_")+"_idx", outputCol=c.replace(" ","_")+"_ohe")
    for c in cat_cols
]
label_indexer = StringIndexer(inputCol="price_range", outputCol="label", handleInvalid="keep")
assembler = VectorAssembler(
    inputCols=[c.replace(" ","_")+"_ohe" for c in cat_cols] + num_cols_rf,
    outputCol="rf_features",
    handleInvalid="skip"
)

preproc_stages = indexers + encoders + [label_indexer, assembler]

# ── Pipeline Random Forest ─────────────────────────────────────────────────
rf = RandomForestClassifier(
    labelCol="label", featuresCol="rf_features",
    numTrees=200, maxDepth=15,
    minInstancesPerNode=5,
    featureSubsetStrategy="sqrt",
    seed=42
)
pipeline_rf = Pipeline(stages=preproc_stages + [rf])
```

```python
# ── Entrenar Random Forest ────────────────────────────────────────────────
model_rf = pipeline_rf.fit(train_df)
pred_rf   = model_rf.transform(test_df)

# Recuperar etiquetas del label para la matriz de confusión
label_im = next(
    s for s in model_rf.stages
    if isinstance(s, StringIndexerModel) and s.getInputCol() == "price_range"
)
label_names = label_im.labelsArray[0]          # ej: ['Mid', 'Low', 'High']
idx_to_label = {float(i): lb for i, lb in enumerate(label_names)}

print("=" * 52)
print("  RANDOM FOREST")
print("=" * 52)
for metric in ["accuracy", "f1"]:
    score = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction",
        metricName=metric).evaluate(pred_rf)
    print(f"  {metric.upper():10s} = {score:.4f}")

# Matriz de confusión con etiquetas legibles
to_label_udf = F.udf(lambda x: idx_to_label.get(x, str(x)))
pred_rf_lbl = (
    pred_rf
    .withColumn("true_label", to_label_udf(col("label")))
    .withColumn("pred_label", to_label_udf(col("prediction")))
)
print("\nMatriz de confusión (RF):")
pred_rf_lbl.groupBy("true_label").pivot("pred_label").count().orderBy("true_label").show()

# Importancia de variables
rf_model     = model_rf.stages[-1]
ohe_names    = []
for c in cat_cols:
    stage = next(
        s for s in model_rf.stages
        if isinstance(s, StringIndexerModel) and s.getInputCol() == c
    )
    lbls = stage.labelsArray[0]
    ohe_names += [f"{c}={lb}" for lb in lbls[:-1]]

feature_names = ohe_names + num_cols_rf
print("\nImportancia de variables (RF):")
for nm, v in sorted(
    zip(feature_names, rf_model.featureImportances.toArray()),
    key=lambda x: -x[1]
):
    print(f"  {nm:<42} {v:.4f}")

# ── Pipeline GBT + OneVsRest ──────────────────────────────────────────────
print("\n" + "=" * 52)
print("  GRADIENT BOOSTED TREES (OneVsRest)")
print("=" * 52)

gbt = GBTClassifier(
    labelCol="label", featuresCol="rf_features",
    maxIter=100, maxDepth=8,
    stepSize=0.1, subsamplingRate=0.8,
    seed=42
)
ovr = OneVsRest(classifier=gbt, labelCol="label", featuresCol="rf_features")
pipeline_gbt = Pipeline(stages=preproc_stages + [ovr])

model_gbt = pipeline_gbt.fit(train_df)
pred_gbt  = model_gbt.transform(test_df)

for metric in ["accuracy", "f1"]:
    score = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction",
        metricName=metric).evaluate(pred_gbt)
    print(f"  {metric.upper():10s} = {score:.4f}")

pred_gbt_lbl = (
    pred_gbt
    .withColumn("true_label", to_label_udf(col("label")))
    .withColumn("pred_label", to_label_udf(col("prediction")))
)
print("\nMatriz de confusión (GBT):")
pred_gbt_lbl.groupBy("true_label").pivot("pred_label").count().orderBy("true_label").show()
```

### 5.5 Interpretación de resultados

El **Random Forest** obtuvo una **Accuracy de 0.6814** y un **F1-Score de 0.6765**. El modelo **GBT (OneVsRest)** superó ligeramente al RF con una **Accuracy de 0.6895** y un **F1-Score de 0.6863**. Ambos valores son sustancialmente superiores al 0.52 que alcanzaría un clasificador trivial que siempre predijera la clase mayoritaria (`Mid`), lo que evidencia que el modelo captura información genuina de las variables predictoras.

El vector de importancias reveló que las variables más determinantes fueron **`Host Listings Count`** (0.2637), seguida de **`Room Type=Entire home/apt`** (0.1222) y **`Room Type=Private room`** (0.0930). ## 6. Modelo de aprendizaje no supervisado (K-Means)

**K-Means** es sencillo y escalable en entornos distribuidos. PySpark MLlib implementa la variante `k-means||`, permite descubrir perfiles latentes de propiedades sin etiquetas, complementando el enfoque supervisado.

### 6.1 Variables utilizadas

Variables numéricas: `Bedrooms`, `Bathrooms`, `Accommodates`, `Beds`, `Cleaning Fee`, `Extra People`, `Guests Included`, `log_reviews`, `Host Listings Count` y `Price`.

### 6.2 Estandarización

Debido a que K-Means opera con distancias euclidianas, se aplica `StandardScaler` para que ninguna variable domine por escala.

### 6.3 Selección del número de clusters mediante método del codo

Se evaluaron valores de $k$ de 2 a 7. El criterio de selección fue el mayor **Silhouette Score**. Los resultados del método del codo fueron:

| k | WSSSE | Silhouette |
|---|-------|-----------|
| 2 | 364,216.88 | **0.6472** ◄ mejor |
| 3 | 331,577.29 | 0.4846 |
| 4 | 297,711.29 | 0.2716 |
| 5 | 269,631.53 | 0.6460 |
| 6 | 240,143.90 | 0.4802 |
| 7 | 237,865.68 | 0.4770 |

El valor de **k = 2** resultó óptimo al maximizar el Silhouette Score (0.6472).

### 6.5 Evaluación

**Silhouette Score** en el rango $[-1, 1]$: valores cercanos a 1 indican clusters bien separados.

```python
# ── Pipeline no supervisado: K-Means ─────────────────────────────────────
# Features ampliadas con las variables de mayor correlación con el precio
cluster_features = [
    "Bedrooms", "Bathrooms", "Accommodates", "Beds",
    "Cleaning Fee", "Extra People", "Guests Included",
    "log_reviews", "Host Listings Count", "Price"
]

assembler_km = VectorAssembler(
    inputCols=cluster_features,
    outputCol="features_raw",
    handleInvalid="skip"
)
scaler = StandardScaler(
    inputCol="features_raw", outputCol="features",
    withMean=True, withStd=True
)

# ── Método del codo: evaluar k = 2…7 ─────────────────────────────────────
print("Método del codo:")
print(f"{'k':>3}  {'WSSSE':>14}  {'Silhouette':>12}")
print("-" * 36)

best_k, best_sil = 3, -1.0
for k in range(2, 8):
    km_tmp = KMeans(k=k, seed=42, featuresCol="features",
                    predictionCol="cluster", maxIter=20)
    pipe_tmp = Pipeline(stages=[assembler_km, scaler, km_tmp])
    m_tmp    = pipe_tmp.fit(muestra_M)
    p_tmp    = m_tmp.transform(muestra_M)
    wssse    = m_tmp.stages[-1].summary.trainingCost
    sil      = ClusteringEvaluator(
        featuresCol="features", predictionCol="cluster",
        metricName="silhouette", distanceMeasure="squaredEuclidean"
    ).evaluate(p_tmp)
    marker = "  ◄ mejor" if sil > best_sil else ""
    if sil > best_sil:
        best_sil = sil
        best_k   = k
    print(f"{k:>3}  {wssse:>14.2f}  {sil:>12.4f}{marker}")

print(f"\nk óptimo (mayor Silhouette): k = {best_k}")

# ── Modelo final con k óptimo ─────────────────────────────────────────────
kmeans_final = KMeans(
    k=best_k, seed=42, featuresCol="features",
    predictionCol="cluster", maxIter=50
)
pipeline_km = Pipeline(stages=[assembler_km, scaler, kmeans_final])
model_km    = pipeline_km.fit(muestra_M)
pred_km     = model_km.transform(muestra_M)

silhouette = ClusteringEvaluator(
    featuresCol="features", predictionCol="cluster",
    metricName="silhouette", distanceMeasure="squaredEuclidean"
).evaluate(pred_km)
print(f"Silhouette Score (k={best_k}) = {silhouette:.4f}")
```

```python
# Perfil descriptivo de cada cluster
pred_km.groupBy("cluster").agg(
    F.count("*").alias("n"),
    F.round(F.mean("Price"),            2).alias("Price_mean"),
    F.round(F.mean("Bedrooms"),         2).alias("Bedrooms_mean"),
    F.round(F.mean("Bathrooms"),        2).alias("Bathrooms_mean"),
    F.round(F.mean("Accommodates"),     2).alias("Accommodates_mean"),
    F.round(F.mean("Beds"),             2).alias("Beds_mean"),
    F.round(F.mean("Cleaning Fee"),     2).alias("CleaningFee_mean"),
    F.round(F.mean("Number of Reviews"),2).alias("Reviews_mean"),
    F.round(F.mean("Host Listings Count"),2).alias("HostListings_mean"),
    F.round(F.mean("Availability 365"), 2).alias("Avail365_mean")
).orderBy("cluster").show()

# Cruce con price_range para validar coherencia con el modelo supervisado
print("Cruce cluster vs. price_range:")
pred_km.groupBy("cluster").pivot("price_range").count().orderBy("cluster").show()
```

### 6.6 Interpretación de los grupos obtenidos

Con **k = 2**, el modelo identificó dos clusters con perfiles claramente diferenciados:

- **Cluster 0 — Económico/Estándar** (n = 39,636): propiedades pequeñas con precio promedio de **$104 USD**, aproximadamente 1 habitación, 1 baño y capacidad para 2-3 personas, con disponibilidad media de 161 días al año. Este cluster concentra la mayoría de propiedades `Low` (11,307) y `Mid` (22,759), correspondiendo al segmento masivo del mercado Airbnb.
- **Cluster 1 — Premium** (n = 8,733): propiedades más grandes con precio promedio de **$296 USD**, aproximadamente 3 habitaciones, 2 baños y capacidad para 6 personas, con tarifa de limpieza media de $101 USD y alta disponibilidad (183 días). Este cluster concentra la mayoría de propiedades `High` (5,936), confirmando coherencia con la segmentación supervisada.

Esta segmentación es coherente con la clasificación supervisada y **proporciona información adicional** sobre la disponibilidad y la dinámica de precios, variables que el etiquetado por precio no captura directamente.

## 7. Discusión de resultados

### 7.1 Desempeño del modelo supervisado

El **Random Forest** alcanzó una Accuracy de **0.6814** y un F1-Score de **0.6765**; el **GBT con OneVsRest** obtuvo una Accuracy de **0.6895** y un F1-Score de **0.6863**, superando ligeramente al RF. Ambos valores son sustancialmente superiores al ~0.52 de un clasificador trivial, evidenciando que las variables estructurales de la propiedad contienen información significativa sobre su rango de precio. El F1-Score, al ser robusto frente al ligero desbalance entre clases, confirma que los modelos no se limitan a predecir la clase mayoritaria, sino que capturan los patrones que diferencian los segmentos Low, Mid y High.

### 7.2 Utilidad práctica de la clasificación de precios

Los modelos de clasificación de precios son útiles en varios escenarios: (i) **recomendación automática de precio** para nuevos anuncios; (ii) **detección de anomalías**, identificando propiedades cuyo precio difiere significativamente del predicho; (iii) **análisis de competencia** mediante comparación con propiedades similares; y (iv) **diseño de políticas de pricing dinámico** por parte de la plataforma.

El Silhouette Score obtenido con **k = 2** fue de **0.6472**, indicando una separación satisfactoria entre los dos clusters. La complementariedad entre el modelo supervisado y el no supervisado enriquece la comprensión del fenómeno.

### 7.4 Aplicaciones para inversión inmobiliaria

Combinando ambos enfoques, un inversionista puede: (i) identificar el **cluster** al que pertenece una propiedad candidata; (ii) estimar el **rango de precio** esperado dadas sus características; (iii) cruzar esta información con la ubicación geográfica para detectar **zonas de alta demanda** (baja disponibilidad anual y alta puntuación); y (iv) priorizar inversiones en perfiles con mayor potencial de retorno.

### 7.5 Limitaciones del estudio

- La muestra del **10%** reduce significativamente el costo computacional pero introduce **varianza muestral** que podría mitigarse con técnicas de remuestreo o validación cruzada.
- La discretización de `Price` en terciles ($p_{25}, p_{75}$) es **arbitraria**: otras particiones (cuartiles, deciles) podrían ofrecer resoluciones distintas.
- No se incorpora información **geoespacial detallada** (más allá del país o ciudad), que probablemente sea uno de los predictores más fuertes del precio.
- No se realizó **búsqueda exhaustiva de hiperparámetros** ni validación cruzada por restricciones de tiempo y cómputo.

### 7.6 Posibles mejoras futuras

2. Probar algoritmos alternativos: **Gradient Boosted Trees** y **Multilayer Perceptron** para clasificación; **Gaussian Mixture** y **Bisecting K-Means** para clustering.
